```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'Temperature': [22, 34, 35, 45, 21, 32, 43, 45, 54, 65],
    'Humidity': [60, 65, 70, 75, 80, 85, 90, 95, 100, 105],
    'Wind Speed': [12, 13, 15, 17, 16, 14, 18, 11, 10, 16],
    'Pressure': [1015, 1012, 1017, 1014, 1013, 1016, 1011, 1018, 1015, 1013]
}

df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
plt.scatter(df['Temperature'], df['Humidity'], color='blue', label='Temp vs Humidity')
plt.title('Temperature vs Humidity', fontsize=14)
plt.xlabel('Temperature (C)', fontsize=12)  # Added missing comma and fontsize
plt.ylabel('Humidity (%)', fontsize=12)  # Added ylabel for clarity
plt.legend()
plt.grid()
plt.show()
```
