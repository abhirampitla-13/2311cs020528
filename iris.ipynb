```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn import preprocessing

# Load the dataset
iris = pd.read_csv('/content/iris.csv',index_col=0)
iris = iris.rename(columns={"Species":"Flowers_type"})

# Encode the target variable
label_encoder = preprocessing.LabelEncoder()
iris['Flowers_type'] = label_encoder.fit_transform(iris['Flowers_type'])

# Split data into features (x) and target (y)
x=iris.iloc[:,0:4]
y=iris.iloc[:,4]

# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)

# Create and train the Decision Tree model
model_gini = DecisionTreeClassifier(criterion='gini', max_depth=3) # Define model_gini here
model_gini.fit(x_train, y_train)

# Make predictions and evaluate the model
preds = model_gini.predict(x_test)
print(preds)
pd.Series(preds).value_counts()

pred=model_gini.predict(x_test)
np.mean(preds==y_test)

# Plot the decision tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(12,8))

tree.plot_tree(model_gini) # model_gini is now accessible
plt.show()
```
