```python
import pandas as pd
df = pd.read_csv("Wholesale customers data.csv")
df.head() # Removed the extra space before df.head()
```

```python
import pandas as pd
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt


db = DBSCAN(eps=0.5, min_samples=5)
y_predicted = db.fit_predict(df[['Fresh', 'Milk']])

df['cluster'] = y_predicted

df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1['Milk'], df1['Fresh'], color='green', label='Cluster 0')
plt.scatter(df2['Milk'], df2['Fresh'], color='red', label='Cluster 1')
plt.scatter(df3['Milk'], df3['Fresh'], color='black', label='Cluster 2')

if -1 in df['cluster'].values:
    noise = df[df.cluster == -1]
    plt.scatter(noise['Milk'], noise['Fresh'], color='yellow', label='Noise')

plt.xlabel('Milk')
plt.ylabel('Fresh')
plt.legend()
plt.title('DBSCAN Clustering')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(df.iloc[:, 1:])
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```
