```python
import pandas as pd
from google.colab import files

# Upload the file containing car data (assuming it's a CSV file)
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Load the data into a DataFrame named 'df'
df = pd.read_csv(file_name) # Changed 'cars' to 'df' to match the variable used later

df.head()
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm # Import the statsmodels library

# Assuming 'df' is your DataFrame and 'WT' and your target variable is 'MPG'
# Define and fit the model
X = df[['KM']] # Independent variable
y = df['HP']  # Dependent variable
X = sm.add_constant(X) # Add a constant to the model
model = sm.OLS(y, X).fit() # Fit the Ordinary Least Squares model

# Now plot the regression results
fig = plt.figure(figsize=(15, 8))
fig = sm.graphics.plot_regress_exog(model, 'KM', fig=fig)
plt.show()
```
